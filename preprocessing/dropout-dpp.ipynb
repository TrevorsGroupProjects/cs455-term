```python
import arcgis
import pandas as pd
from arcgis.gis import GIS
from arcgis.features import FeatureLayer
```

```python
dropout_rate_url = "https://services.arcgis.com/P3ePLMYs2RVChkJx/arcgis/rest/services/ACS_Youth_Activity_Boundaries/FeatureServer/1"
```

```python
feature_layer_drop_outs = FeatureLayer(dropout_rate_url, gis)
```

```python
unique_state_keys = feature_layer_drop_outs.query(where="1=1", out_fields=["State"], return_distinct_values=True, return_geometry=False)
leave_offs = ["Puerto Rico", "District of Columbia"]
states_in_data = [state.as_dict["attributes"]["State"] for state in list(unique_state_keys) if state.as_dict["attributes"]["State"] not in leave_offs]
states_in_data
```

```python
queries_by_state = [f"State = '{state}'" for state in states_in_data]
queries_by_state
```

```python
fields = ["NAME", "State", "B14005_001E", "B14005_002E", "B14005_008E", "B14005_012E", "B14005_013E", "B14005_016E", "B14005_022E", "B14005_026E", "B14005_027E"]
```

```python
data_frames=[feature_layer_drop_outs.query(where=query_by_state, out_fields=fields, as_df=True, return_geometry=False) for query_by_state in queries_by_state]
len(data_frames)
```

```python
usa_data_frame = pd.DataFrame(columns=fields)
for data_frame in data_frames:
    usa_data_frame = pd.concat([usa_data_frame, data_frame])
usa_data_frame
```
