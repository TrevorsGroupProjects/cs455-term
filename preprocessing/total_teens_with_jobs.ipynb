```python
import arcgis
import pandas as pd
from arcgis.gis import GIS
from arcgis.features import FeatureLayer
```

```python
labor_force_participation_url = "https://services.arcgis.com/P3ePLMYs2RVChkJx/arcgis/rest/services/ACS_Labor_Force_Participation_by_Age_Boundaries/FeatureServer/1"
```

```python
feature_layer_teens_with_jobs = FeatureLayer(labor_force_participation_url, gis)
```

```python
unique_state_keys = feature_layer_teens_with_jobs.query(where="1=1", out_fields=["State"], return_distinct_values=True, return_geometry=False)
leave_offs = ["Puerto Rico", "District of Columbia"]
states_in_data = [state.as_dict["attributes"]["State"] for state in list(unique_state_keys) if state.as_dict["attributes"]["State"] not in leave_offs]
states_in_data
```

```python
queries_by_state = [f"State = '{state}'" for state in states_in_data]
queries_by_state
```

```python
fields = ["NAME", "State", "B23027_002E", "B23027_003E"]
```

```python
data_frames=[feature_layer_teens_with_jobs.query(where=query_by_state, out_fields=fields, as_df=True, return_geometry=False) for query_by_state in queries_by_state]
len(data_frames)
```

```python
usa_data_frame = pd.DataFrame(columns=fields)
for data_frame in data_frames:
    usa_data_frame = pd.concat([usa_data_frame, data_frame])
usa_data_frame
```
