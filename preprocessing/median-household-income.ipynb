```python
unique_state_keys = feature_layer_median_household_income.query(where="1=1", out_fields=["State"], return_distinct_values=True, return_geometry=False)
leave_offs = ["Puerto Rico", "District of Columbia"]
states_in_data = [state.as_dict["attributes"]["State"] for state in list(unique_state_keys) if state.as_dict["attributes"]["State"] not in leave_offs]
states_in_data
```

```python
queries_by_state = [f"State = '{state}'" for state in states_in_data]
queries_by_state
```

```python
column_page_size = 5
iterations = math.ceil(len(fields)/column_page_size)
iterations
```

```python
pages = []

#Chop up all of the columns present in the feature layer and query all the rows of a state in column sections
for i in range(iterations):
    previous_index = i * column_page_size
    end = column_page_size + previous_index
        
    #print(f"({previous_index}, {end})")
    pages.append(fields[previous_index:end])

pages
```

```python
data_frames_by_state=[feature_layer_median_household_income.query(where="State = 'Alabama'", out_fields=page, as_df=True, return_geometry=True) for page in pages]
len(data_frames_by_state)
```

```python
def merge_data_frames(state_data_frames):
    initial_frame = state_data_frames[0]
    data_frames_by_state = state_data_frames[0:]
    for df in state_data_frames:
        initial_frame = initial_frame.merge(df)
    return initial_frame
```

```python
merge_data_frames(data_frames_by_state)
```

```python
data_frames = []
for query in queries_by_state:
    data_frames_by_state=[feature_layer_median_household_income.query(where=query, out_fields=page, as_df=True, return_geometry=False) for page in pages]
    data_frames.append(data_frames_by_state)
len(data_frames)
```

```python
print(len(data_frames[0]))
```

```python
merged_state_dfs = [merge_data_frames(dfs) for dfs in data_frames]    
merged_state_dfs
```

```python
for merged_state in merged_state_dfs:
    state = merged_state["State"][0]
    file_name = f"Median_Household_Income_{state}.csv"
    merged_state.to_csv(file_name)
```

```python
usa_data_frame = pd.DataFrame(columns=fields)
for data_frame in merged_state_dfs:
    usa_data_frame = pd.concat([usa_data_frame, data_frame])
usa_data_frame
```
