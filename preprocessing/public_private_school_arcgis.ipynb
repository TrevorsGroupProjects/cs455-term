```python
import arcgis
import pandas as pd
from arcgis.gis import GIS
from arcgis.features import FeatureLayer
```

```python
gis = arcgis.gis.GIS(username="rm3468", password="DyAccu1DtNpl4Mn8Ejtu")
gis.url
```

```python
public_school_feature_layer = FeatureLayer(public_school_url, gis)
private_school_feature_layer = FeatureLayer(private_school_url, gis)
```

```python
unique_state_keys = public_school_feature_layer.query(where="1=1", out_fields=["STATE"], return_distinct_values=True, return_geometry=False)
leave_offs = ["DC", "PR", "GU", "VI", "AS"]
states_in_data = [state.as_dict["attributes"]["STATE"] for state in list(unique_state_keys) if state.as_dict["attributes"]["STATE"] not in leave_offs]
```

```python
queries_by_state = [f"State = '{state}'" for state in states_in_data]
queries_by_state
```

```python
all_public_schools = [public_school_feature_layer.query(where=query_by_state, as_df=True) for query_by_state in queries_by_state]
len(all_public_schools)
```

```python
all_private_schools = [private_school_feature_layer.query(where=query_by_state, as_df=True, return_count_only=False) for query_by_state in queries_by_state]
len(all_private_schools)
```

```python
public_school_columns = all_public_schools[0].columns
private_school_columns = all_private_schools[0].columns
```

```python
public_school_dataframe = pd.DataFrame(columns=public_school_columns)
for data_frame in all_public_schools:
    public_school_dataframe = pd.concat([public_school_dataframe, data_frame])
public_school_dataframe
```

```python
private_school_dataframe = pd.DataFrame(columns=private_school_columns)
for data_frame in all_private_schools:
    private_school_dataframe = pd.concat([private_school_dataframe, data_frame])
private_school_dataframe
```

```python
public_school_dataframe.to_csv("usa_public_schools.csv")
private_school_dataframe.to_csv("usa_private_schools.csv")
```
